```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python

data=pd.read_csv('/content/drive/MyDrive/ML Datasets/student_performance_dataset.csv')
data.head()
```

```python
encoding_map = {'Low': 0, 'Medium': 1, 'High': 2}
data['performance'] = data['performance'].replace(encoding_map)
data
```

```python
X=data.iloc[:,0:3]
X.head()
```

```python
y=data.iloc[:,3]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# **LOGISTIC REGRESSION FOR MULTICLASS CLASSIFICATION**

```python
logistic=LogisticRegression(multi_class='ovr')
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
y_pred
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

# **TIME FOR SOME HYPERPARAMETER TUNING**
